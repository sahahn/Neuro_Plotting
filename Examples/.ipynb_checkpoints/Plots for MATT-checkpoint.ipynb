```python
import pandas as pd
import matplotlib.gridspec as gridspec
import numpy as np
import matplotlib.pyplot as plt

from Neuro_Plotting.Ref import SurfRef, VolRef, load_mapping, save_mapping
from Neuro_Plotting.Plot import Plot_Surf, Plot_Surf_Collage, Plot_SurfVol_Collage
from nilearn.plotting import plot_glass_brain, plot_stat_map
```

```python
rois = pd.read_csv('/home/sage/Neuro_Plotting/Examples/Data for Figure.csv').dropna()
rois = rois.drop(['Beta.SE_Zero', 'Unnamed: 0'], axis=1)
```

```python
surf_ref = SurfRef(space='fsaverage', parc='desikan')
```

```python
thickness = surf_ref.get_hemis_plot_vals(rois, 'lh_', 'rh_', i_keys=['thickness'])
area = surf_ref.get_hemis_plot_vals(rois, 'lh_', 'rh_', i_keys=['area'])
```

```python
# Most of these params are default, just showing more than needed as an example
figure, axes, smf = Plot_Surf_Collage(data = thickness,
                                      ref = surf_ref,
                                      surf_mesh = 'inflated',
                                      bg_map = 'sulc',
                                      view = 'standard',
                                      cmap = 'cold_hot',
                                      title = 'Thickness -  Beta Weights',
                                      title_sz = 22,
                                      vmax = .02,
                                      vmin = -.02,
                                      threshold = .0005,
                                      wspace = 0,
                                      hspace = 0,
                                      figsize = (15, 10),
                                      midpoint = None,
                                      avg_method = 'median',
                                      bg_on_data = True,
                                      alpha = 1,
                                      colorbar = True)

plt.savefig('Thickness_Hires.png', dpi=300, bbox_inches='tight')
plt.savefig('Thickness_Lowres.png', dpi=100, bbox_inches='tight')
```

```python
figure, axes, smf = Plot_Surf_Collage(data = area,
                                      ref = surf_ref,
                                      surf_mesh = 'inflated',
                                      bg_map = 'sulc',
                                      view = 'standard',
                                      cmap = 'cold_hot',
                                      title = 'Surface Area - Beta Weights',
                                      title_sz = 22,
                                      vmax = .00008,
                                      vmin = -.00008,
                                      threshold = .000005,
                                      wspace = 0,
                                      hspace = 0,
                                      figsize = (15, 10),
                                      midpoint = None,
                                      avg_method = 'median',
                                      bg_on_data = True,
                                      alpha = 1,
                                      colorbar = True)

plt.savefig('Area_View1_Hires.png', dpi=300, bbox_inches='tight')
plt.savefig('Area_View1_Lowres.png', dpi=100, bbox_inches='tight')
```

```python
figure, axes, smf = Plot_Surf_Collage(data = area,
                                      ref = surf_ref,
                                      surf_mesh = 'inflated',
                                      bg_map = 'sulc',
                                      view = 'fb',
                                      cmap = 'cold_hot',
                                      title = '  Surface Area - Beta Weights',
                                      title_sz = 22,
                                      vmax = .00008,
                                      vmin = -.00008,
                                      threshold = .000005,
                                      wspace = -.4,
                                      hspace = 0,
                                      figsize = (15, 5),
                                      midpoint = None,
                                      avg_method = 'median',
                                      bg_on_data = True,
                                      alpha = 1,
                                      colorbar = True)

plt.savefig('Area_View2_Hires.png', dpi=300, bbox_inches='tight')
plt.savefig('Area_View2_Lowres.png', dpi=100, bbox_inches='tight')
```

```python
vol_ref = VolRef(space='mni', parc='aseg')
volume = vol_ref.get_plot_vals(rois, i_keys=['_volume'])
```

```python
plot_glass_brain(volume, cmap='cold_hot', 
                 vmax=.00003,
                 threshold=0.000001,
                 symmetric_cbar=True,
                 plot_abs=False,
                 colorbar=True,
                 title='Subcortical Volume - Beta Weights')

plt.savefig('Glass_Brain_Volume_Hires.png', dpi=300, bbox_inches='tight')
plt.savefig('Glass_Brain_Volume_Lowres.png', dpi=100, bbox_inches='tight')
```
